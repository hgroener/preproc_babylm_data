```python
def preproc(text):
    t = html.unescape(text)
    t = re.sub("&amp;", "&", t)
    t = re.sub("&lt;.*?&gt;", "", t)
    t = re.sub("&lt;", "", t)
    t = re.sub("&gt;", "", t)
    t = re.sub("\[.*?\]", "", t)
    t = re.sub("\(.*?\)", "", t)
    t = re.sub("<.*?>", "", t)
    t = re.sub("-LSB-.*?-RSB-", "", t)
    t = re.sub("(_BOOK_TITLE_ : .*\.txt)(\.out)?", "", t)
    t = re.sub("-[A-Z]+-|_[A-Z_]+_", "", t)
    t = re.sub("CHAPTER[A-Z ]+", " ", t)
    t = re.sub("\n.*?\d\|\|\d.*?\n", "\n", t)
    t = re.sub("\n[\* ]+\n", "\n", t)
    t = re.sub("\t", "", t)
    
    
    
    t = re.sub("\n\n+", "\n", t)
    t = re.sub(" [ ]+", " ", t)
    
    return(t)

    
```
